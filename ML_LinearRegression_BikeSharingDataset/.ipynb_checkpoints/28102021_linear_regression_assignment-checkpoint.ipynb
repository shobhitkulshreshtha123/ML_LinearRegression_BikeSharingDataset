<b> Columns Understanding </b>

Following information about various columns has been extracted from data dictionary. Columns further understanding is highlighted by the researcher wherever an explanation is needed:

- instant: record index `This column acts just as a serial number and is thus redundant. It will be dropped in next step.`


- dteday : date `This column will be used to extract date for trend analysis. Year will not be extracted since they are already present in subequent columns.`
    - Date could be useful in identifying the dates on which boombikes are in most and least demand (Apart from holiday and day of the week)


- season : season (1:spring, 2:summer, 3:fall, 4:winter) `This feature will be converted to names of the season for later usage so that proper dummies can be created`


- yr : year (0: 2018, 1:2019)

 
- mnth : month ( 1 to 12) `This feature will be mapped to month name and will be later converted to dummy if found relevant for the model.`


- holiday : weather day is a holiday or not (extracted from http://dchr.dc.gov/page/holiday-schedule)


- weekday : day of the week `This feature will be mapped to the day name and will be later converted to dummy if found relevant for the model.`


- workingday : if day is neither weekend nor holiday is 1, otherwise is 0.


- weathersit : `Numeric values in this variable will be converted to below mentioned text entries. Post conversion, patterns will be visualized and dummy variables will be created before the model development`
	- 1: Clear, Few clouds, Partly cloudy, Partly cloudy `This will be converted to "Clear to Partly Cloudy"`
	- 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist `This will be converted to "misty and cloudy"`
	- 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds `This will be converted to "light rain and light snow"`
	- 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog `This will be converted to "heavy rain and heavy snow"`


- temp : temperature in Celsius


- atemp: feeling temperature in Celsius


- hum: humidity


- windspeed: wind speed


- casual: count of casual users `Unregistered customers`


- registered: count of registered users


- cnt: count of total rental bikes including both casual and registered `This is the Target Variable. Since this column is a sum of casual and registered, we will drop casual and registered in next step (Being redundant)

<b>Variable Mappin as explained earlier:</b>

```python
# Mapping weathersit feature:

boombikes['weathersit']= boombikes['weathersit'].map({1: 'Clear to Partly Cloudy', 2: 'Misty and Cloudy', 
                                                      3: 'Light Rain or Snow', 4: 'Heavy Rain or Snow'})
```

```python
# Mapping Season feature:

boombikes['season']= boombikes['season'].map({1:'Spring', 2:'Summer', 3:'Fall', 4:'Winter'})
```

```python
# Mapping Month feature:

boombikes['mnth']= boombikes['mnth'].map({1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
                      7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'})
```

```python
# Mapping Weekday feature:

boombikes['weekday']= boombikes['weekday'].map({0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'})
```

# Data Visualization

## Univariate Analysis

### Numerical Variables - Related to Target Variable

#### Boxplot & Distplot Visualization

Although holiday, workingday, season, year, mnth, weekday, weathersit, date are shown as numerical values, they actually are categorical in nature. Hence, we will exclude these while creating boxplots for univariate analysis.

```python
# Creating a list of features for boxplot visualization:

uni_num= ['casual', 'registered', 'cnt']
```

```python
# Creating a subplot grid
fig, ax= plt.subplots(nrows= 1, ncols= len(uni_num), figsize= (12, 5), sharey= True)

# Placing main title
plt.suptitle('boxplots showing basic customer trends'.upper(), fontsize= 20)

# Looping to fill subplot grid with plots
for i in range(len(uni_num)):
    
    sns.boxplot(y=boombikes[uni_num[i]], ax= ax[i], palette= 'Wistia')
    
    # Setting aesthetics and readability
    ax[i].set_title(f'{uni_num[i].upper()} CUSTOMERS', fontsize= 15)
    ax[i].set_ylabel(' ')

# Setting final aesthetics
plt.tight_layout()
plt.show()
```

<b>Observations:</b>
- Casual customers have shown presence of outliers. This indicates that on certain days relatively more number of casual customers have used Boombikes. However, relatively more values are packed between Q1 to Q3 showing that the distribution if right skewed.
- Number of casual customers is far lesser then registered customers. This signifies that registered customers are using 
- spread of resistered customers and total customers is somewhat even signifying a possible normalized distribution.

```python

# Creating a subplot grid
fig, ax= plt.subplots(nrows= 1, ncols= len(uni_num), figsize= (12, 5))

# Placing main title
plt.suptitle('distplots showing customer type distribution'.upper(), fontsize= 20)

# Looping to fill subplot grid with plots
for i in range(len(uni_num)):
    
    sns.distplot(boombikes[uni_num[i]], ax= ax[i], color= 'brown')
    
    # Setting aesthetics and readability
    ax[i].set_title(f'{uni_num[i].upper()} CUSTOMERS', fontsize= 15)
    ax[i].set_ylabel(' ')
    ax[i].set_xlabel(' ')

# Setting final aesthetics
plt.tight_layout()
plt.show()
```

<b>Observations:</b>
- Above distplots prove our intuitive understanding:
    -Distribution of Casual customers is right skewed 
    
<b>Data follows normal distribution of values of registered customers and target variable.</b>

## Bivariate Analysis

### Numerical Variables

We will consider folloing features for this analysis: temp, atemp, hum, windspeed

```python
# Deriving list of features

biv_num= ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered']
```

```python

# Creating a subplot grid
fig, ax= plt.subplots(nrows= 2, ncols= int(len(biv_num)/2), figsize= (20, 9), sharey= True)

# Placing main title
plt.suptitle('relationship of continuous variables with target variable'.upper(), fontsize= 20)

# Looping to fill subplot grid with plots
k= 0
for i in range(2):
    for j in range(int(len(biv_num)/2)):
        sns.regplot(x= boombikes[biv_num[k]], y=boombikes['cnt'], ax= ax[i, j], color= 'red', 
                    scatter_kws= {'color': 'yellow'}, line_kws= {'color': 'red'})
    
        # Setting aesthetics and readability
        ax[i,j].set_title(f'{biv_num[k].upper()}', fontsize= 15)
        ax[i,j].set_ylabel(' ')
        ax[i,j].set_xlabel(' ')
    
        k+=1

# Setting final aesthetics
plt.tight_layout()
plt.show()
```

<b> Observations: </b>
- Demand seems to be directly proportional with temperature and feeling temperature. i.e, the customer count increases with increase in termperature and feeling temperature.
- Demand has an inversely proportional relationship with humidity and windspeed. i.e, customer count decreases with increase in windspeed and humidity.
- Relationship of total customer count with registered users is much stronger then that of the customer count with casual users.

### Categorical Variables

We will consider following variables for this analysis: season, yr, mnth, holiday, weekday, workingday, weathersit, date

```python
# Deriving list of features

biv_cat= ['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit', 'date']
```

```python

# Creating a subplot grid
fig, ax= plt.subplots(nrows= 2, ncols= int(len(biv_cat)/2), figsize= (20, 12))

# Placing main title
plt.suptitle('relationship of categorical variables with target variable'.upper(), fontsize= 20)

# Looping to fill subplot grid with plots
k= 0
for i in range(2):
    for j in range(int(len(biv_cat)/2)):
        sns.barplot(y= boombikes.groupby(by= biv_cat[k]).sum()['cnt'], 
                    x= boombikes.groupby(by= biv_cat[k]).sum().index, ax= ax[i,j])
    
        # Setting aesthetics and readability
        ax[i,j].set_title(f'{biv_cat[k].upper()}', fontsize= 15)
        ax[i,j].set_ylabel(' ')
        ax[i,j].set_xlabel(' ')
    
        k+=1

# Setting final aesthetics
plt.tight_layout()
plt.show()
```

<b>Observations:</b>
- Maximum demand is observed in fall season. Within Fall season, highest demand is noticed in September with a slight decrease in October followed by steep increase in November.
- Summer season witnesses second highest demand. Within summer season, April shows lowest demand followed by steep rise in demand in the month of May followed by another steep increase in June.
- Winter season witnesses third highest demand with highest in the month of December followed by steep decrease in January followed by further decline in February.
- Spring season has shown least demand. Demand stays less in February but shows stark rise in March.
- Friday, Saturday, Sunday and Thursday are high demand days.
- 60% increase in number of users in 2019 as comapred to that of in 2018.
